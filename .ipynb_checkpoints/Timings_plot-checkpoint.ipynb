```python
%matplotlib widget
import numpy as np
import matplotlib.pyplot as plt
import scipy.special
import os
import h5py
```

```python
corrsteps = 5
nt = 1024
k = 5

# read in output data
def read_Int_timings(output_file):
    file = open(output_file).readlines()
    numlines = len(file)
    print(numlines, (nt-k-1)*(corrsteps+1))

    data = np.zeros((nt-k-1, corrsteps+1, 3))

    for l,line in enumerate(file):
        t = l//(corrsteps+1)
        c = l%(corrsteps+1)
        nums = line.split()
        data[t,c,0] = float(nums[0])
        data[t,c,1] = float(nums[1])
        data[t,c,2] = float(nums[2])  
        
    return data

data = read_Int_timings("./New_vs_Old_timings.dat")
```

```python
plt.figure()
for i in np.arange(corrsteps+1):
    plt.plot(np.arange(nt-k-1), data[:,i,0], label = "old" if i==corrsteps else "")
    plt.plot(np.arange(nt-k-1), data[:,i,1], label = "new blocks" if i==corrsteps else "")
    plt.plot(np.arange(nt-k-1), data[:,i,2], label = "new corr" if i==corrsteps else "")
    plt.plot(np.arange(nt-k-1), data[:,i,1]+data[:,i,2], label = "new total" if i==corrsteps else "")
plt.legend()
plt.yscale("log")
plt.xscale("log")
```
